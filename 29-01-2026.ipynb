# T-Test

**Definition:** T-Test is used to compare the means when the population varience is unknown or the sample size is small.

## Independent T-Test

**Example:**  compare **Glucose** of **Positive vs Negtive** person.

**Hypotheses:**

H₀ : Glucose distribution of positive = Glucose distribution of negative

H₁ : Glucose distribution of positive ≠ Glucose distribution of negative

```python

pregnancies_pos = df[df['Outcome'] == 1]['Pregnancies']
pregnancies_negative = df[df['Outcome'] == 0]['Pregnancies']


t_stat, p_val = stats.ttest_ind(pregnancies_pos,pregnancies_negative)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
if p_val < 0.05:
    print("Reject null hypothesis")
else:
    print("accept null hypothesis")
```
